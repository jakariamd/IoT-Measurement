```python
import csv
import os
import pandas as pd

# Post Process OUI
csv_path = "../../../Validation Data/UNSW/CSV-OUI/"

# Use os.listdir to get a list of files in the folder
files = os.listdir(csv_path)

global_frame = pd.DataFrame(columns=['mac', 'oui', 'oui_friendly'])

for file in files:
    try:
        # Open the CSV file and read line by line
        data = []
        with open(csv_path+file, 'r', newline='') as f:
            reader = csv.reader(f)
            for row in reader:
                data.append([row[0], row[1], ", ".join(set(row[2:]))])

        data = pd.DataFrame(data).dropna().drop_duplicates()
        data.columns = ['mac', 'oui', 'oui_friendly']
        data = data[data['oui'] != '']
        data = data.groupby(['mac', 'oui'])['oui_friendly'].agg(lambda x: ", ".join(set(', '.join(x).split(', ')))).reset_index()
        global_frame = pd.concat([global_frame, data], ignore_index=True)
    except:
        print(f"{file} : empty file" )
# 
global_frame.drop_duplicates(inplace=True)
global_frame = global_frame.groupby(['mac', 'oui'])['oui_friendly'].agg(lambda x: ", ".join(set(', '.join(x).split(', ')))).reset_index()
global_frame = global_frame.drop_duplicates('mac')

# todo save result in file 
# global_frame.to_csv('../../../Validation Data/UNSW/scrape-results/oui_friendly.csv')
```
